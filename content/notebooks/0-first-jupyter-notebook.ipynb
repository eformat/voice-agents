## 4. Check for GPU

Let's verify that a GPU is available in this workbench using `nvidia-smi`.

```python
import subprocess

result = subprocess.run(["nvidia-smi"], capture_output=True, text=True)
if result.returncode == 0:
    print(result.stdout)
else:
    print("No GPU detected - nvidia-smi not available")
```

## 5. Working with lists and loops

A simple example to practice running cells.

```python
toppings = ["pepperoni", "mushrooms", "olives", "basil", "mozzarella"]

for i, topping in enumerate(toppings, 1):
    print(f"{i}. {topping}")

print(f"\nTotal toppings: {len(toppings)}")
```

## 6. Writing a function

Try creating your own function.

```python
def pizza_order(size, toppings):
    base_price = {"small": 8, "medium": 12, "large": 16}
    price = base_price.get(size, 12) + len(toppings) * 1.5
    return f"{size.title()} pizza with {', '.join(toppings)} - ${price:.2f}"

print(pizza_order("large", ["pepperoni", "mushrooms", "olives"]))
print(pizza_order("small", ["mozzarella"]))
```
